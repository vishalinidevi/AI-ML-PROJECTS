```python
import requests
import openai
from langchain import OpenAI, PromptTemplate, LLMChain

# Function to fetch the web page content
def fetch_web_content(url):
    response = requests.get(url)
    response.raise_for_status()
    return response.text

# Function to summarize the content using OpenAI via LangChain
# def summarize_content(content, api_key):
#     openai.api_key = api_key
#     llm = OpenAI(api_key=api_key)
#     prompt_template = PromptTemplate(
#         input_variables=["content"],
#         template="Summarize the following content:\n\n{content}"
#     )
#     # Create the prompt from the template
#     prompt = prompt_template.format(content=content)
#     chain = LLMChain(llm=llm, prompt=prompt) # Pass 'prompt' instead of 'prompt_template'
#     summary = chain.run(content=content)
#     return summary

def summarize_content(content, api_key):
    openai.api_key = api_key
    llm = OpenAI(api_key=api_key)
    prompt_template = PromptTemplate(
        input_variables=["content"],
        template="Summarize the following content:\n\n{content}"
    )
    # Pass 'prompt_template' instead of 'prompt' to LLMChain
    chain = LLMChain(llm=llm, prompt=prompt_template)
    summary = chain.run(content=content)
    return summary

# Function to generate an answer based on the summary and a question
def generate_answer(question, summary, api_key):
    openai.api_key = api_key
    llm = OpenAI(api_key=api_key)
    prompt_template = PromptTemplate(
        input_variables=["summary", "question"],
        template="Based on the following summary:\n\n{summary}\n\nAnswer the question: {question}"
    )
    chain = LLMChain(llm=llm, prompt_template=prompt_template)
    answer = chain.run(summary=summary, question=question)
    return answer

# Main code
url = "https://www.rcktechiees.com/why-rck"
api_key = "you-api-key"

# Fetch the content
content = fetch_web_content(url)

# Summarize the content
summary = summarize_content(content, api_key)
print("Summary:", summary)

# Generate answer to a question
question = "What is the main topic of the website?"
answer = generate_answer(question, summary, api_key)
print("Answer:", answer)
```

```python
import requests
from transformers import pipeline

# Function to fetch the web page content
def fetch_web_content(url):
    response = requests.get(url)
    response.raise_for_status()
    return response.text

# Function to chunk the text
def chunk_text(text, chunk_size=1000, overlap=200):
    chunks = []
    start = 0
    while start < len(text):
        end = min(start + chunk_size, len(text))
        chunk = text[start:end]
        chunks.append(chunk)
        start += chunk_size - overlap
    return chunks

# Function to summarize the content using Hugging Face transformers
def summarize_content(content):
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    chunks = chunk_text(content)
    summaries = [summarizer(chunk, max_length=150, min_length=30, do_sample=False)[0]['summary_text'] for chunk in chunks]
    return " ".join(summaries)

# Function to generate an answer based on the summary and a question
def generate_answer(question, summary):
    qa_model = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
    answer = qa_model(question=question, context=summary)
    return answer['answer']

# Main code
url = "https://fobegolf.com/about-us.html"  # Example URL, replace with your URL
content = fetch_web_content(url)

# Summarize the content
summary = summarize_content(content)
print("Summary:", summary)

# Generate answer to a question
question = "What is the purpose of fobegolf?"
answer = generate_answer(question, summary)
print("Answer:", answer)
```
